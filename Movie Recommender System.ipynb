```python
#for converting string to list as we want geres only like (Adventure, Fantasy) nothing more than that
import ast
```

```python
def convert_g(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
#Here we got the list of all the genres for every movie
movies['genres'] = movies['genres'].apply(convert_g)
```

```python
#Now we have to do it with the cast as well, but we want to get only top 3 actors
def convert_c(obj):
    l = []
    c = 0
    for i in ast.literal_eval(obj):
        if c != 3:
            l.append(i['name'])
            c += 1
        else :
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_c)
```

```python
movies['cast']
# Now here we got the 3 main actors for every movie
```

```python
#Now we focus on the column crew to fetch the name of director which is important for recommendation
movies.crew[0]
```

```python
def fetch_d(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_d)
```

```python
#Now we will do stemming as we saw that the words like act, action.actions have similiar meaning
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y= []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
#Now repeating the above steps for vectorinsation
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words ='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
#Now we can see that repeatation is not there so we have solved that problem
```

```python
#Measurement of co sine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
#It is showing the cosine distance of cosine between every movie
similarity.shape
```

```python
#this is giving the cosine distance of movie at index 0 to each and every movie
similarity[0]
```

```python
#With the below code we are able to sort the cosine distances but we are loosing the index position of the movie
sorted(similarity[0],reverse=True)
```

```python
#For holding the index position of the movie as its very important we use enumerate funation, 
#used lambda in key as we wanted to sort by cosine distance not the indexs that enumerate has created
#Fetch top 5 out of these as we want top five movies
sorted(list(enumerate((similarity[0]))),reverse = True, key = lambda x:x[1])[1:6]
```

```python
#Finally we make a function which is giving us top five movies which should be recommended depending upon the search
def recommendation(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index] 
    movie_list = sorted(list(enumerate((distances))),reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movie_list:
        x = new_df.iloc[i[0]]['title']
        print(x)
    
```

```python
recommendation('Batman Begins')
```
